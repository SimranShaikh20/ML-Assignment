#  Step 1: Load the Boston Housing Dataset from CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset from a CSV file (update the path as needed)
df = pd.read_csv('/content/HousingData.csv')  # Replace with your actual file path

# Quick check of the dataset
print(df.head())
print(df.columns)
```

 # 1️⃣ Boxplot for the 'Median value of owner-occupied homes' (MEDV)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['MEDV'], color="skyblue")
plt.title('Boxplot of Median Value of Owner-Occupied Homes (MEDV)')
plt.xlabel('MEDV ($1000s)')
plt.show()
```

# 2️⃣ Histogram for the Charles River variable (CHAS)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df['CHAS'], palette="Set2")
plt.title('Histogram of Charles River Variable (CHAS)')
plt.xlabel('Bounds Charles River (0 = No, 1 = Yes)')
plt.ylabel('Frequency')
plt.show()
```

# 3️⃣ Boxplot of MEDV vs Discretized AGE Variable

```python
# Discretize AGE into 3 groups
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=[0, 35, 50, 100],
                         labels=['<=35', '36-50', '>50'])

plt.figure(figsize=(8, 6))
sns.boxplot(x='AGE_GROUP', y='MEDV', data=df, palette='pastel')
plt.title('Boxplot of MEDV vs AGE Group')
plt.xlabel('Age Group')
plt.ylabel('MEDV ($1000s)')
plt.show()
```

# 4️⃣ Scatter Plot: NOX vs INDUS

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='INDUS', y='NOX', data=df, color='coral')
plt.title('Scatter Plot of NOX vs INDUS')
plt.xlabel('Proportion of Non-retail Business Acres (INDUS)')
plt.ylabel('Nitric Oxide Concentration (NOX)')
plt.show()
```
