```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your car dataset CSV file (adjust path)
df = pd.read_csv('/content/car_dataset_india.csv')
```

```python
print(df.columns)
print(df.head())
```

```python
display(df)
```

# Drop rows with missing values in key numeric columns

```python

df_clean = df.dropna(subset=['Year', 'Price', 'Mileage', 'Engine_CC', 'Seating_Capacity', 'Service_Cost'])
```

```python

# Convert columns to correct numeric types if needed
numeric_cols = ['Year', 'Price', 'Mileage', 'Engine_CC', 'Seating_Capacity', 'Service_Cost']
for col in numeric_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

# 1. ðŸ“ˆ Line Chart: Price vs Year

```python


plt.figure(figsize=(10, 5))
plt.plot(df_clean['Year'], df_clean['Price'], 'o', alpha=0.5)
plt.title('Car Price vs Year')
plt.xlabel('Year')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.show()
```

# 2. ðŸ“Š Histogram: Mileage Distribution


```python

plt.figure(figsize=(8, 5))
plt.hist(df_clean['Mileage'], bins=30, color='skyblue', edgecolor='black')
plt.title('Mileage Distribution')
plt.xlabel('Mileage (km/l)')
plt.ylabel('Number of Cars')
plt.show()
```

# 3. ðŸ”— Correlation Matrix


```python

plt.figure(figsize=(10, 6))
corr = df_clean[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

# 4. ðŸ” Multivariate Plot (Pairplot)


```python

sns.pairplot(df_clean[numeric_cols])
plt.suptitle("Pairwise Relationship Between Numeric Features", y=1.02)
plt.show()
```

# 5. ðŸ¥§ Pie Chart: Seating Capacity Distribution
-

```python

seat_counts = df_clean['Seating_Capacity'].value_counts().sort_index()
plt.figure(figsize=(7, 7))
plt.pie(seat_counts, labels=seat_counts.index.astype(str), autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Seating Capacity")
plt.axis('equal')
plt.show()
```
